# Q3.

## 95% CI is an interval that covers the middle 95% of the bootstrapped means, which means that we can use np.quantile(...) to calculate the 0.25% and 97.5% quartiles of the bootstrapped means to find out the confidence interval. 

# Q4.

````python
import numpy as np

# Sample data (replace with your own data)
data = np.array([12, 15, 14, 10, 13, 17, 18, 16, 11, 19])  # Example data

# Number of bootstrap samples
n_bootstraps = 1000

# Create an array to store the means of the bootstrap samples
boot_means = []

# Generate bootstrap samples and calculate the mean for each
for _ in range(n_bootstraps):
    # Resample from the original data with replacement
    bootstrap_sample = np.random.choice(data, size=len(data), replace=True)
    # Calculate the mean of the bootstrap sample
    boot_means.append(np.mean(bootstrap_sample))

# Convert the list of bootstrapped means to a numpy array
boot_means = np.array(boot_means)

# Calculate the 2.5th and 97.5th percentiles using np.quantile
lower_bound = np.quantile(boot_means, 0.025)
upper_bound = np.quantile(boot_means, 0.975)

# Output the 95% confidence interval
print(f"95% bootstrap confidence interval for the population mean: [{lower_bound}, {upper_bound}]")
````

### Summary of Exchanges

**1. Standard Error of the Mean (SEM) vs. Standard Deviation (SD)**:
- **SD** measures the spread or variability of individual data points in the original dataset.
- **SEM** measures the precision of the sample mean as an estimate of the true population mean.
- **SEM** is calculated by dividing the SD by the square root of the sample size: \(\text{SEM} = \frac{\text{SD}}{\sqrt{n}}\).

**2. Creating a 95% Confidence Interval from SEM**:
- The 95% confidence interval (CI) estimates the range within which the true population mean likely lies.
- The formula for the 95% CI is: 
  \[
  \text{CI} = \bar{x} \pm 1.96 \times \text{SEM}
  \]
  where 1.96 is the critical value from the standard normal distribution for 95% confidence.

**3. What is a Confidence Interval and Why 95% Falls Under 1.96 Standard Deviations**:
- A **confidence interval** provides a range for the population parameter estimate based on sample data.
- For a 95% confidence level, 95% of the data in a normal distribution falls within 1.96 standard deviations from the mean.
  
**4. Using `np.quantile(...)` on Bootstrapped Sample Means to Get a 95% Confidence Interval**:
- To compute the 95% confidence interval using bootstrapped sample means:
  - Generate multiple bootstrapped samples, calculate the mean for each.
  - Use `np.quantile()` to find the **2.5th** and **97.5th** percentiles of the bootstrapped sample means, which provide the lower and upper bounds of the confidence interval:
    \[
    \text{CI} = [\text{np.quantile}(boot_means, 0.025), \text{np.quantile}(boot_means, 0.975)]
    \]

## It is critical to distinguish population parameter and sample statistics when it comes to confidence intervals because the purpose of confidence intervals is to estimate the unknown population parameter from the sample statistics drawn from the population. 

## Thus, although confidence intervals provide an estimation for population parameter, it can not be overconfidently regarded as the population parameter as it is just an estimate from the sample statistics. 

# Q6.

## What is the process of bootstrapping?
    The process of bootstrapping is by using the original sample of size n, we create a hypothetical new bootstrapped sample by drawing entries n times with replacement from the original sample. Then, we record the statistics of the new sample and repeat the procedure.

## What is the main purpose of bootstrapping?
    The main purpose of bootstrapping is to replicate the sampling multiple times to create an estimate of the population that we don't know.

## If you had a (hypothesized) guess about what the average of a population was, and you had a sample of size n from that population, how could you use bootstrapping to assess whether or not your (hypothesized) guess might be plausible?
    We can bootstrap samples of size n multiple times and calculate the means of the bootstrapped samples. Recorded bootstrapped samples' means will be in a form of a histogram. Lastly, by using the 95% confidence interval, we can estimate whether or not my guess might be plausible by checking if my guess is in the interval or not.

# Q7.

## Why does a confidence interval overlapping zero "fail to reject the null hypothesis" when the observed sample mean statistic itself is not zero?
    Confidence interval overlapping zero means that zero is one of the plausible values for the population parameter. This is not a strong evidence to reject the null hypothesis because zero is in the confidence interval and is a plausible value for the population parameter. 
    
## Alternatively, what would lead to the opposite conclusion in this context; namely, instead choosing "to reject the null hypothesis"?
    To reject the null hypothesis, the confidence intervals must not include zero. If the confidence interval does not include zero, then zero cannot be a plausible value for the population parameter, and thus, it is a strong evidence to reject the null hypothesis.

## We need to find whether the Vaccine is effective or not. Thus, I have added a new column named "Effect of Vaccine" and used "FinalHealthScore" - "InitialHealthScore" for the data.

## Positive values for "Effect of Vaccine" column means that the vaccine has positive effects on health and negative values means that the vaccine has negative effects on health. 0 for the column means that the final health score and the initial health score is equal, and thus, the vaccine does nothing, which is our null hypothesis.

````python
import pandas as pd

# Replace 'your_local_path' with the path where you saved the CSV file
url = 'https://github.com/BillyWoongjaeCheon/STA130---Assignments/raw/e6265210c73d884355b2a9c3844df647b33fa772/patient_health_data.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(url)

# Calculate the "Effect of Vaccine" and add it as a new column
df['Effect of Vaccine'] = df['FinalHealthScore'] - df['InitialHealthScore']

df
````

## Then, I bootstrapped from this sample 1000 times to create an estimate of the population mean. I recorded the mean of each bootstrapped samples into a histogram and calculated the mean of the bootstrapped samples and the 95% confidence interval.

````python
import numpy as np
import plotly.express as px

# Number of bootstrap repetitions
n_iterations = 1000

# Array to hold the means from each bootstrap sample
bootstrap_means = []

# Bootstrap sampling
for _ in range(n_iterations):
    # Generate a bootstrap sample
    bootstrap_sample = df['Effect of Vaccine'].sample(frac=1, replace=True)
    
    # Calculate the mean of the bootstrap sample and store it
    bootstrap_means.append(bootstrap_sample.mean())

# Convert the list to a NumPy array for easier analysis
bootstrap_means = np.array(bootstrap_means)

# Calculate the confidence intervals
lower_ci = np.percentile(bootstrap_means, 2.5)
upper_ci = np.percentile(bootstrap_means, 97.5)

# Print the results
print(f"Bootstrap Mean: {np.mean(bootstrap_means)}")
print(f"95% Confidence Interval: [{lower_ci}, {upper_ci}]")

# Create a histogram using Plotly
fig = px.histogram(
    x=bootstrap_means,
    nbins=30,
    title='Bootstrap Distribution of Effect of Vaccine Means',
    labels={'x': 'Mean Effect of Vaccine', 'y': 'Frequency'},
)

# Add lines for mean and confidence intervals
fig.add_vline(x=np.mean(bootstrap_means), line_color='blue', line_width=2, line_dash='dash', annotation_text='Mean', annotation_position='top right')
fig.add_vline(x=lower_ci, line_color='red', line_width=2, line_dash='dash', annotation_text='2.5% CI', annotation_position='top right')
fig.add_vline(x=upper_ci, line_color='green', line_width=2, line_dash='dash', annotation_text='97.5% CI', annotation_position='top right')

# Show the figure
fig.show(renderer="png")
````

## As the confidence interval does not overlap with 0, the null hypothesis, we have a strong evidence to reject the null hypothesis as 0 is not a plausible value for the population mean.
